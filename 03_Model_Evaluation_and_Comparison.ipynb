### Flight Delay Prediction: Model Evaluation and Comparison

This notebook presents a comprehensive graphic evaluation and comparison of the machine learning models employed for predicting flight delays. The models include baseline regression models, untuned ensemble methods, and tuned ensemble methods. We will visualize the performance of these models using error bars to represent the mean Root Mean Squared Error (RMSE) and its standard deviation. This analysis aims to identify the most effective model for flight delay prediction and understand the impact of hyperparameter tuning on model performance.


```python
# Import libraries
import matplotlib.pyplot as plt
```

```python
# Function to plot model evaluations
def plot_model_evaluation(title, list_model, list_mean, list_std, ylim=(18.8, 21)):
    plt.figure(figsize=(8, 6))
    for i in range(len(list_model)):
        plt.errorbar(i, list_mean[i], yerr=list_std[i], fmt='o', linewidth=3, capsize=5, capthick=2)
    plt.xticks(range(len(list_model)), list_model)
    plt.title(title, fontsize=15)
    plt.ylim(ylim)
    plt.grid(True, color='#666666', linestyle='-')
    plt.minorticks_on()
    plt.grid(True, color='#999999', linestyle='-', alpha=0.2)
    plt.ylabel('Minutes', fontsize=12)
    plt.xlabel('Model', fontsize=12)
    plt.show()
```

```python
### Baseline regression models
# Evaluation data
baseline_models = ['Linear Regression', 'Decision Tree', 'KNN']
baseline_means = [20.673, 19.606, 19.742]
baseline_stds = [0.161, 0.15, 0.219]

# Plot baseline model evaluation
plot_model_evaluation("Baseline Model Evaluation", baseline_models, baseline_means, baseline_stds)

### Insights:
# The Decision Tree model outperforms the Linear Regression and KNN models in terms of lower mean RMSE.
# All baseline models have relatively low standard deviations, indicating consistent performance across cross-validation folds.
```

```python
### Ensemble methods (without tuning)
# Evaluation data
ensemble_models_untuned = ['Random Forest', 'Boosting (Adaboost)', 'Bagging']
ensemble_means_untuned = [19.854, 19.468, 19.948]
ensemble_stds_untuned = [0.208, 0.156, 0.216]

# Plot ensemble methods evaluation (without tuning)
plot_model_evaluation("Ensemble Methods Evaluation (Without Tuning)", ensemble_models_untuned, ensemble_means_untuned, ensemble_stds_untuned)

### Insights:
# Among the untuned ensemble methods, Adaboost shows the best performance with the lowest mean RMSE.
# Both Boosting and Bagging outperform the untuned Random Forest model.
```

```python
### Ensemble methods (with tuning)
# Evaluation data
ensemble_models_tuned = ['Random Forest', 'Boosting (Adaboost)', 'Bagging']
ensemble_means_tuned = [19.048, 19.45, 19.319]
ensemble_stds_tuned = [0.155, 0.145, 0.152]

# Plot ensemble methods evaluation (with tuning)
plot_model_evaluation("Ensemble Methods Evaluation (With Tuning)", ensemble_models_tuned, ensemble_means_tuned, ensemble_stds_tuned)

### Insights:
# After hyperparameter tuning, the Random Forest model shows a significant improvement and achieves the lowest mean RMSE among the ensemble methods.
# The standard deviations are slightly reduced for the tuned models, indicating more consistent performance.
# Hyperparameter tuning improves the performance of all ensemble methods, with Random Forest benefiting the most.
```
